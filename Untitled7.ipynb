**LOADER**


```python
from langchain_community.document_loaders.pdf import PyPDFLoader
from pathlib import Path
from pprint import pp
```

```python
file_path = Path("/content/sample_data/Internship_Report.pdf")

file_path.exists()
```

```python
pypdf_loader = PyPDFLoader(file_path=file_path.as_posix(),
                           mode='page')
```

```python
documents = pypdf_loader.load()
```

```python
len(documents)
```

```python
pp(documents[0].page_content)
```

**SPLITTER/CHUNKING**


```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
texts = text_splitter.split_documents(documents)
```

```python
print(f"Total chunks created: {len(texts)}")
```

**EMBEDDINGS**

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS


embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")


vector_store = FAISS.from_documents(texts, embeddings)

print("Successfully embedded all chunks!")
```

```python
vector_store.index.ntotal
```

```python

user_question = "What is biodiversity"

matched_docs = vector_store.similarity_search(user_question, k=3)

print(f"Retrieved {len(matched_docs)} chunks from the PDF:\n")
for i, doc in enumerate(matched_docs):
    print(f"--- Chunk {i+1} (Page {doc.metadata.get('page')}) ---")
    print(f"{doc.page_content[:200]}...\n")
```

**EXTRACTION**

```python
import os
from langchain_groq import ChatGroq

from langchain_core.tools import create_retriever_tool

from langchain.agents import create_agent as create_react_agent

import os
from google.colab import userdata

os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')


retriever = vector_store.as_retriever(search_kwargs={"k": 6})
pdf_search_tool = create_retriever_tool(
    retriever,
    name="pdf_knowledge_base",
    description="Use this tool to search for specific context or facts inside the uploaded PDF document."
)
tools = [pdf_search_tool]

llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0)

agent_executor = create_react_agent(llm, tools)

response = agent_executor.invoke({"messages": [{"role": "user", "content": "What are the core findings mentioned in the PDF?"}]})

print("\nFinal Output:")
print(response["messages"][-1].content)
```

```python

your_question = "Look specifically at Chapter 1, Section 1.1 'Objective of the Internship' on Page 9. What was the primary objective?"

response = agent_executor.invoke({
    "messages": [{"role": "user", "content": your_question}]
})

print("\nFinal Output:")
print(response["messages"][-1].content)
```

```python
from langchain_core.tools import tool
from langchain.agents import create_agent

@tool
def read_pdf_page_content(page_number: int) -> str:
    """Use this tool to read the raw text of a specific page number directly.
    Pass 1 for Page 1, 2 for Page 2, etc."""
    try:
        return documents[page_number - 1].page_content
    except IndexError:
        return f"Error: Page {page_number} does not exist."

tools = [pdf_search_tool, read_pdf_page_content]

agent_executor = create_agent(
    model=llm,
    tools=tools
)

print("Agent built successfully with both tools!")
```
